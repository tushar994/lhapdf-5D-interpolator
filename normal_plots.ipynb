# Normal plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def custom_func_np(ln_x, ln_q):
    x_val = np.exp(1.5/( ( 0.1 * ((ln_x + 7)**2) ) + 1) )/(np.log(np.exp(ln_q) + 0.007) + 5) + np.exp(-1*(ln_x+12)/5) - 1
    return x_val
```

```python
data = pd.read_csv("./custom_func_data/filename4.csv")
data
```

```python
data['function'] = custom_func_np(data['lnx'], data['lnq'])
```

```python
data['ratio'] = data['xf']/data['function']
```

```python
data['interpolator_value'] = data['xf']
```

```python
graphing = [data[i:i+1800].reset_index() for i in range(0,341999,1800)]
graphing
```

```python
i = 0
for dat in graphing:
    i+=1
    if(i==10):
        i=0
        print("ln(q) is " , (dat['lnq'][0]))
        dat.plot(x ='lnx', y=['interpolator_value','function'])
        plt.show()
```
