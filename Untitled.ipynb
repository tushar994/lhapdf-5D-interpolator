```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def custom_func_np(ln_x, ln_q):
    x_val = np.exp(1.5/( ( 0.1 * ((ln_x + 7)**2) ) + 1) )/(np.log(np.exp(ln_q) + 0.007) + 5) + np.exp(-1*(ln_x+12)/5) - 1
    return x_val
```

```python
data = pd.read_csv("./dense_grids/6-points.csv")
data['x'] = np.exp(data['lnx'])
data['first-derivative-x'] = np.gradient(data['xf'],data['x'])
data['second-derivative-x'] = np.gradient(data['first-derivative-x'],data['x'])
data['function-val'] = custom_func_np(data['lnx'],data['lnq'])
data['first-derivative-x-function'] = np.gradient(data['function-val'],data['x'])
data['second-derivative-x-function'] = np.gradient(data['first-derivative-x-function'],data['x'])
data['first-d-ratio'] = data['first-derivative-x']/data['first-derivative-x-function']
data['second-d-ratio'] = data['second-derivative-x']/data['second-derivative-x-function']
data['value_ratio'] = data['xf']/data['function-val']
```

```python
graphing = [data[i:i+18001].reset_index() for i in range(0,342019,18001)]
```

# Plots for interpolator that uses 6 points

# Value

```python

for graph in graphing:
    if(graph['lnq'][0]==0.5 or graph['lnq'][0]==3.5 or graph['lnq'][0]==6.5 or graph['lnq'][0]==9.5):
        print("for lnq = ", graph['lnq'][0])
        graph.plot(x='lnx',y=['xf','function-val'])
        graph.plot(x='lnx',y=['value_ratio'])
        plt.ylim(0.8,1.2)
        plt.show()
        i=0
```
